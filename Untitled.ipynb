```python
import numpy as np
import pandas as pd
```

```python
import os
ROOT_DIR = './data/2019/07/31'
filenames = []
names = []
for r,d,f in os.walk(ROOT_DIR):
    if d is not None:
        for company in d:
            names.append(company)

    for file in f:
        if file.endswith('.csv'):
            filenames.append(r+'/'+file)

df_list = [pd.read_csv(filename, delimiter=";", engine='python', encoding='utf-8').sort_index(axis=0, ascending=False).reset_index(drop=True) for filename in filenames]
```

```python
df_list = list(map(lambda df: df[df['Vol.'] != '-']['Price'],df_list))
```

```python
df_list = list(map(lambda df: (df-df.min())/(df.max()-df.min()), df_list))
```

```python
distance_matrix = np.zeros((9,9))
```

```python
from tslearn.metrics import dtw

for i in range(9):
    for j in range(9):
        distance_matrix[i,j] = dtw(df_list[i], df_list[j])
```

```python
from scipy import stats

for i in range(9):
    for j in range(9):
        distance_matrix[i,j] = stats.ks_2samp(df_list[i].values, df_list[j].values)[0]
```
